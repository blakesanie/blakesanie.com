```python
import requests

def get_strava_streams(activity_id: int, access_token: str):
    base_url = "https://www.strava.com/api/v3"
    endpoint = f"/activities/{activity_id}/streams"
    keys = "latlng,time,distance,heartrate,velocity_smooth,watts,cadence"
    
    url = f"{base_url}{endpoint}"
    params = {
        "keys": keys,
        "key_by_type": "true"
    }
    headers = {
        "Authorization": f"Bearer {access_token}"
    }

    response = requests.get(url, headers=headers, params=params)

    if response.status_code != 200:
        raise Exception(f"Error fetching streams: {response.status_code} - {response.text}")

    return response.json()

# streams = get_strava_streams(14968454406, res['access_token'])
# streams
```
