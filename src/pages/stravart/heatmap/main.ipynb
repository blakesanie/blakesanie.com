```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the first image
globalHeatmap = np.array(Image.open("global.png"))

# Load the second image
myHeatmap = np.array(Image.open("mine.png"))

blank = np.array(Image.open("blank.png"))
```

```python
plt.imshow(myHeatmap)
```

```python
plt.imshow(globalHeatmap)
```

```python
diff = globalHeatmap - myHeatmap
diff[:,:,-1] = 255
diff
```

```python
plt.imshow(blank + diff)
```

```python
Image.fromarray(blank + abs(diff)).save("diff.png")
```

```python
mask = myHeatmap[:, :, 0] > blank[:, :, 0]
```

```python
gh = globalHeatmap.copy()
gh[mask] = blank[mask]
Image.fromarray(gh).save("out.png")
plt.imshow(gh)
```
