```python
import requests
from datetime import datetime
```

```python
code = "117bd99e036d3d81de1c7c9f45a2d1ad7e17c22a"
```

```python
# URL for the token request
url = "https://www.strava.com/oauth/token"

# Data to be sent in the form
data = {
    'client_id': '132542',
    'client_secret': '2027f2a31dfc685382bb11187e17b32cfc827d4e',
    'code': code,
    'grant_type': 'authorization_code'
}

# Making the POST request
response = requests.post(url, data=data)

# Print the response (for example, the access token)
res = response.json()
res
```

```python
# URL for the GET request
activities_responses = []

# Headers
headers = {
    'accept': 'application/json',
    'Authorization': 'Bearer ' + res['access_token']
}

for i in range(10):

    url = f'https://www.strava.com/api/v3/athlete/activities?per_page=100&page={i+1}'

    # Making the GET request
    response = requests.get(url, headers=headers)

    # Print the response (JSON format)
    activities_res = response.json()
    activities_responses.append(activities_res)

activities_responses
```

```python
activities = []
for response in activities_responses:
    for activity in response:
        activities.append(activity)
len(activities)
```

```python
import json


with open('activities.json', 'w') as f:
    f.write(json.dumps(activities))
```

```python
# load list of activity dicts into pandas df
import pandas as pd

df = pd.DataFrame(activities)
df.head()
```

```python
# filter to get rows where 'type' is 'Ride'
df_rides = df[df['type'] == 'Ride']
df_rides.head()
```

```python
df_2025 = df_rides[df_rides['start_date_local'].str.contains('2025')]
df_2025['edit_url'] = 'https://www.strava.com/activities/' + df_2025['id'].astype(str) + '/edit'
df_2025['avg_mph'] = df_2025['average_speed'] * 2.23694
df_2025['start_date_parsed'] = pd.to_datetime(df_2025['start_date_local']).dt.strftime('%Y-%m-%d %H:%M:%S')
df_2025['miles'] = df_2025['distance'] / 1000 * 0.62
df_2025['hours'] = df_2025['moving_time'] / 3600
```
